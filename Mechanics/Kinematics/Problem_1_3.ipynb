# Problem 1.3: Car Motion Analysis

## Problem Statement
A car starts moving rectilinearly, first with acceleration w = 5.0 m/s² (the initial velocity is equal to zero), then uniformly, and finally, decelerating at the same rate w, comes to a stop. The total time of motion equals t = 25 s. The average velocity during that time is equal to ⟨v⟩ = 72 km per hour. How long does the car move uniformly?

## Mathematical Analysis

Let's break down the problem:

- The car moves in three phases:
  1. Accelerating with $w = 5.0$ m/s² from rest
  2. Moving uniformly (constant velocity)
  3. Decelerating with $w = 5.0$ m/s² until stopping

- The total time of motion is $t = 25$ s
- The average velocity is $\langle v \rangle = 72$ km/h = $72 \times \frac{1000}{3600} = 20$ m/s

Let's denote:
- $t_1$ = time of acceleration phase
- $t_2$ = time of uniform motion phase (this is what we need to find)
- $t_3$ = time of deceleration phase

We know that $t_1 + t_2 + t_3 = 25$ s.

During the acceleration phase, the velocity increases from 0 to some maximum value $v_{max}$. Since the acceleration is constant at $w = 5.0$ m/s², we have:

$v_{max} = w \cdot t_1 = 5.0 \cdot t_1$ m/s

During the deceleration phase, the velocity decreases from $v_{max}$ to 0 with the same magnitude of acceleration (but negative). Since the rate is the same, and we're going from $v_{max}$ to 0 (instead of 0 to $v_{max}$), the time taken for deceleration equals the time taken for acceleration:

$t_3 = t_1$

So we have $t_1 + t_2 + t_1 = 25$, which gives us $t_2 = 25 - 2t_1$.

Now, to find $t_1$ (and consequently $t_2$), we'll use the condition that the average velocity is 20 m/s.

The average velocity is the total distance divided by the total time:

$\langle v \rangle = \frac{S_{total}}{t_{total}} = \frac{S_1 + S_2 + S_3}{t_1 + t_2 + t_3} = 20$ m/s

Where:
- $S_1$ = distance covered during acceleration
- $S_2$ = distance covered during uniform motion
- $S_3$ = distance covered during deceleration

For the acceleration phase with constant acceleration from rest:
$S_1 = \frac{1}{2} \cdot w \cdot t_1^2 = \frac{1}{2} \cdot 5.0 \cdot t_1^2 = 2.5 \cdot t_1^2$ m

For the uniform motion phase with velocity $v_{max}$:
$S_2 = v_{max} \cdot t_2 = 5.0 \cdot t_1 \cdot t_2$ m

For the deceleration phase (which is symmetric to the acceleration phase):
$S_3 = S_1 = 2.5 \cdot t_1^2$ m

So, the total distance is:
$S_{total} = S_1 + S_2 + S_3 = 2.5 \cdot t_1^2 + 5.0 \cdot t_1 \cdot t_2 + 2.5 \cdot t_1^2 = 5.0 \cdot t_1^2 + 5.0 \cdot t_1 \cdot t_2$ m

Substituting $t_2 = 25 - 2t_1$, we get:
$S_{total} = 5.0 \cdot t_1^2 + 5.0 \cdot t_1 \cdot (25 - 2t_1) = 5.0 \cdot t_1^2 + 125 \cdot t_1 - 10.0 \cdot t_1^2 = 125 \cdot t_1 - 5.0 \cdot t_1^2$ m

Now, using the average velocity condition:
$\langle v \rangle = \frac{S_{total}}{t_{total}} = \frac{125 \cdot t_1 - 5.0 \cdot t_1^2}{25} = 20$ m/s

Solving for $t_1$:
$\frac{125 \cdot t_1 - 5.0 \cdot t_1^2}{25} = 20$

$125 \cdot t_1 - 5.0 \cdot t_1^2 = 20 \cdot 25 = 500$

$5.0 \cdot t_1^2 - 125 \cdot t_1 + 500 = 0$

Using the quadratic formula:
$t_1 = \frac{125 \pm \sqrt{125^2 - 4 \cdot 5.0 \cdot 500}}{2 \cdot 5.0} = \frac{125 \pm \sqrt{15625 - 10000}}{10} = \frac{125 \pm \sqrt{5625}}{10} = \frac{125 \pm 75}{10}$

So, $t_1 = \frac{125 + 75}{10} = 20$ s or $t_1 = \frac{125 - 75}{10} = 5$ s.

Since $t_1$ represents the time of the acceleration phase, it must be positive and less than the total time of 25 s. Both solutions satisfy this, but we need to check which one gives a valid value for $t_2$.

For $t_1 = 20$ s:
$t_2 = 25 - 2 \cdot 20 = 25 - 40 = -15$ s, which is negative and thus invalid.

For $t_1 = 5$ s:
$t_2 = 25 - 2 \cdot 5 = 25 - 10 = 15$ s, which is positive and thus valid.

Therefore, the car moves uniformly for $t_2 = 15$ s.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, Math
```

```python
# Define the parameters
w = 5.0  # acceleration/deceleration in m/s²
t_total = 25.0  # total time in seconds
v_avg = 20.0  # average velocity in m/s (72 km/h)

# Solve the quadratic equation: 5.0*t_1^2 - 125*t_1 + 500 = 0
a = 5.0
b = -125.0
c = 500.0

# Calculate the discriminant
discriminant = b**2 - 4*a*c
print(f"Discriminant = {discriminant}")

# Calculate the two solutions
t1_sol1 = (-b + np.sqrt(discriminant)) / (2*a)
t1_sol2 = (-b - np.sqrt(discriminant)) / (2*a)

print(f"Solution 1: t₁ = {t1_sol1:.1f} s")
print(f"Solution 2: t₁ = {t1_sol2:.1f} s")

# Calculate t₂ for both solutions
t2_sol1 = t_total - 2*t1_sol1
t2_sol2 = t_total - 2*t1_sol2

print(f"For t₁ = {t1_sol1:.1f} s, t₂ = {t2_sol1:.1f} s (invalid as it's negative)")
print(f"For t₁ = {t1_sol2:.1f} s, t₂ = {t2_sol2:.1f} s (valid solution)")

# Use the valid solution
t1 = t1_sol2
t2 = t2_sol2
t3 = t1  # Deceleration time equals acceleration time

# Calculate maximum velocity
v_max = w * t1
print(f"Maximum velocity = {v_max:.1f} m/s = {v_max*3.6:.1f} km/h")

# Calculate distances for each phase
s1 = 0.5 * w * t1**2
s2 = v_max * t2
s3 = 0.5 * w * t3**2
s_total = s1 + s2 + s3

print(f"Distance during acceleration: {s1:.1f} m")
print(f"Distance during uniform motion: {s2:.1f} m")
print(f"Distance during deceleration: {s3:.1f} m")
print(f"Total distance: {s_total:.1f} m")

# Verify average velocity
v_avg_calculated = s_total / t_total
print(f"Calculated average velocity: {v_avg_calculated:.1f} m/s = {v_avg_calculated*3.6:.1f} km/h")
print(f"Given average velocity: {v_avg:.1f} m/s = {v_avg*3.6:.1f} km/h")
```

```python
# Create a static visualization of the motion
def create_motion_visualization(t1, t2, t3, w):
    # Time points
    t_total = t1 + t2 + t3
    t = np.linspace(0, t_total, 1000)
    
    # Maximum velocity
    v_max = w * t1
    
    # Initialize arrays for position, velocity, and acceleration
    position = np.zeros_like(t)
    velocity = np.zeros_like(t)
    acceleration = np.zeros_like(t)
    
    # Fill arrays based on the phase of motion
    for i, time in enumerate(t):
        if time <= t1:  # Acceleration phase
            acceleration[i] = w
            velocity[i] = w * time
            position[i] = 0.5 * w * time**2
        elif time <= t1 + t2:  # Uniform motion phase
            acceleration[i] = 0
            velocity[i] = v_max
            position[i] = 0.5 * w * t1**2 + v_max * (time - t1)
        else:  # Deceleration phase
            time_in_phase = time - (t1 + t2)
            acceleration[i] = -w
            velocity[i] = v_max - w * time_in_phase
            position[i] = 0.5 * w * t1**2 + v_max * t2 + v_max * time_in_phase - 0.5 * w * time_in_phase**2
    
    # Create a figure with three subplots
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 12), sharex=True)
    
    # Position vs Time plot
    ax1.plot(t, position, 'b-', linewidth=2)
    ax1.set_ylabel('Position (m)')
    ax1.set_title('Position vs Time')
    ax1.grid(True, alpha=0.3)
    
    # Add phase boundaries and annotations
    ax1.axvline(x=t1, color='r', linestyle='--', alpha=0.7)
    ax1.axvline(x=t1+t2, color='r', linestyle='--', alpha=0.7)
    
    # Annotate the phases
    ax1.text(t1/2, position.max()*0.8, 'Acceleration', ha='center')
    ax1.text(t1 + t2/2, position.max()*0.8, 'Uniform Motion', ha='center')
    ax1.text(t1 + t2 + t3/2, position.max()*0.8, 'Deceleration', ha='center')
    
    # Velocity vs Time plot
    ax2.plot(t, velocity, 'g-', linewidth=2)
    ax2.set_ylabel('Velocity (m/s)')
    ax2.set_title('Velocity vs Time')
    ax2.grid(True, alpha=0.3)
    
    # Add horizontal line for average velocity
    v_avg = position[-1] / t_total
    ax2.axhline(y=v_avg, color='k', linestyle='--', alpha=0.7, label=f'Average Velocity = {v_avg:.1f} m/s')
    ax2.legend()
    
    # Add phase boundaries
    ax2.axvline(x=t1, color='r', linestyle='--', alpha=0.7)
    ax2.axvline(x=t1+t2, color='r', linestyle='--', alpha=0.7)
    
    # Acceleration vs Time plot
    ax3.plot(t, acceleration, 'r-', linewidth=2)
    ax3.set_xlabel('Time (s)')
    ax3.set_ylabel('Acceleration (m/s²)')
    ax3.set_title('Acceleration vs Time')
    ax3.grid(True, alpha=0.3)
    
    # Add phase boundaries
    ax3.axvline(x=t1, color='r', linestyle='--', alpha=0.7)
    ax3.axvline(x=t1+t2, color='r', linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    plt.suptitle(f'Car Motion Analysis: t₁ = {t1:.1f} s, t₂ = {t2:.1f} s, t₃ = {t3:.1f} s', fontsize=16, y=0.98)
    
    return fig

# Create and display the static visualization with the calculated values
fig = create_motion_visualization(t1, t2, t3, w)
plt.show()
```

```python
# Create an interactive visualization
print("Interactive Visualization (only works when running the notebook):")

# Create sliders for parameters
w_slider = widgets.FloatSlider(value=5.0, min=1.0, max=10.0, step=0.1, description='w (m/s²):')
t_total_slider = widgets.FloatSlider(value=25.0, min=10.0, max=40.0, step=1.0, description='Total Time (s):')
v_avg_slider = widgets.FloatSlider(value=20.0, min=5.0, max=30.0, step=1.0, description='Avg. Velocity (m/s):')

# Function to update the visualization
def update_visualization(w, t_total, v_avg):
    # Solve the quadratic equation for t1
    a = w/2
    b = -w*t_total/2
    c = v_avg*t_total
    
    discriminant = b**2 - 4*a*c
    
    if discriminant < 0:
        print("No real solutions exist for these parameters.")
        return
    
    t1_sol1 = (-b + np.sqrt(discriminant)) / (2*a)
    t1_sol2 = (-b - np.sqrt(discriminant)) / (2*a)
    
    # Calculate t2 for both solutions
    t2_sol1 = t_total - 2*t1_sol1
    t2_sol2 = t_total - 2*t1_sol2
    
    # Choose the valid solution (t2 must be positive)
    if t2_sol1 > 0:
        t1 = t1_sol1
        t2 = t2_sol1
    elif t2_sol2 > 0:
        t1 = t1_sol2
        t2 = t2_sol2
    else:
        print("No valid solutions exist for these parameters.")
        return
    
    t3 = t1  # Deceleration time equals acceleration time
    
    # Calculate maximum velocity
    v_max = w * t1
    
    # Calculate distances for each phase
    s1 = 0.5 * w * t1**2
    s2 = v_max * t2
    s3 = 0.5 * w * t3**2
    s_total = s1 + s2 + s3
    
    # Verify average velocity
    v_avg_calculated = s_total / t_total
    
    # Display the results
    print(f"Solution: t₁ = {t1:.1f} s, t₂ = {t2:.1f} s, t₃ = {t3:.1f} s")
    print(f"Maximum velocity = {v_max:.1f} m/s = {v_max*3.6:.1f} km/h")
    print(f"Calculated average velocity: {v_avg_calculated:.1f} m/s = {v_avg_calculated*3.6:.1f} km/h")
    
    # Create the visualization
    create_motion_visualization(t1, t2, t3, w)

# Create interactive output
interactive_output = widgets.interactive_output(update_visualization, 
                                               {'w': w_slider, 't_total': t_total_slider, 'v_avg': v_avg_slider})

# Display widgets and output
display(widgets.VBox([w_slider, t_total_slider, v_avg_slider, interactive_output]))
```

We have solved Problem 1.3 and found that the car moves uniformly for **15 seconds**.

The complete motion of the car can be broken down as follows:
1. Acceleration phase: 5 seconds (from 0 to 25 m/s)
2. Uniform motion phase: 15 seconds (at 25 m/s)
3. Deceleration phase: 5 seconds (from 25 m/s to 0)

The total distance traveled is 500 meters, and the average velocity is 20 m/s (72 km/h), which matches the given conditions.

The visualizations above illustrate how the position, velocity, and acceleration of the car change over time, providing a clear picture of the three phases of motion.
